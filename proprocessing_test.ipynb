```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
name = '../behavior/behavior1/IMG/center_2017_12_16_11_42_44_965.jpg'
# name = '../behavior/shadow.jpg'

plt.figure()
image = cv2.imread(name)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)

plt.figure()
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap="gray")

plt.figure()
image_eh = cv2.equalizeHist(image_gray)
plt.imshow(image_eh, cmap="gray")

plt.figure()
image_re = cv2.resize(image_gray, (64, 64), interpolation=cv2.INTER_AREA)
plt.imshow(image_re, cmap="gray")
```

```python
# Horizontal Flip
plt.figure()
image = cv2.imread(name)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb = cv2.flip(image_rgb, 1)
plt.imshow(image_rgb)
```

```python
# Horizontal transform
plt.figure()
tx = 150
ty = 0
image = cv2.imread(name)
rows,cols,_ = image.shape
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
M = np.float32([[1,0,tx],[0,1,ty]])
image_rgb = cv2.warpAffine(image_rgb,M,(cols,rows))
plt.imshow(image_rgb)


left_name = '../behavior/behavior1/IMG/left_2017_12_16_11_42_44_965.jpg'
plt.figure()
image = cv2.imread(left_name)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```
